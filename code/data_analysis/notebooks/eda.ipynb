```python
def own_stat(df):
    columns = df.columns
    column_data = []
    length = df.shape[0]
    
    for column in columns:
        data = {}
        data["name"] = column
        data["type"] = df[column].dtypes
        data["n_unique"] = df[column].nunique()
        data["na"] = df[column].isna().sum()
        data["na_percentage"] = round(data["na"] / length * 100,4) 
        column_data.append(data)
    
    result_df = pd.DataFrame(column_data)
    return result_df
```

```python
own_stat(df)
```
